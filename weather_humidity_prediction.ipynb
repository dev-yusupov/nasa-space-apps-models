```python
import requests
import pandas as pd

# Define the NASA POWER API endpoint
url = "https://power.larc.nasa.gov/api/temporal/daily/point"

# Define parameters for the API call
params = {
    'start': '20000101',  # Start date (YYYYMMDD)
    'end': '20240930',    # End date (YYYYMMDD)
    'latitude': 38.861,   # Latitude of Tajikistan (can be adjusted)
    'longitude': 71.276,  # Longitude of Tajikistan (can be adjusted)
    'community': 'ag',    # Community (agriculture)
    'parameters': 'RH2M', # Relative humidity at 2 meters
    'format': 'JSON',     # Data format (JSON for easy handling in Python)
    'header': 'true',     # Include metadata in the response
}

# Make a GET request to the API
response = requests.get(url, params=params)

# Parse the JSON response
data = response.json()

# Extract the relevant data (daily relative humidity)
humidity_data = data['properties']['parameter']['RH2M']

# Convert the data into a Pandas DataFrame for analysis
df = pd.DataFrame(list(humidity_data.items()), columns=['Date', 'Relative Humidity (%)'])

# Display the first few rows of the data
df
```
